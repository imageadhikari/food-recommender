```python
# Load the saved model
loaded_model = load_model("savedmodel/")
```

```python
# Taking user inputs

# dow = input("Enter the week from 0 to 6")
# hour_of_day = input("Enter the hour of day from 0 to 23")

dow = 4
hour_of_day = 14
```

```python
dict = {
    'product_id': [],
    'order_dow': [],
    'order_hour_of_day': [],
    'aisle_id':[],
    'department_id': []
}
```

```python
for key, (value1, value2) in loaded_dictionary.items():
    dict['product_id'].append(key)
    dict['order_dow'].append(dow)
    dict['order_hour_of_day'].append(hour_of_day)
    dict['aisle_id'].append(value1)
    dict['department_id'].append(value2)
```

```python
for key in dict:
    dict[key] = np.array(dict[key]).astype('int64')
```

```python
# Use the loaded model for predictions
pred_ans = loaded_model.predict(dict, batch_size=256)
```

```python
print(pred_ans)
```

#### Predicted items

```python
# Selecting indices of the items with highest recommendation values
# Flatten the array to get a 1-dimensional array
flattened_arr = pred_ans.flatten()

# Get the indices that would sort the array in descending order
sorted_indices = np.argsort(flattened_arr)[::-1]

# Get the top 10 indices
top_10_indices = sorted_indices[:10]

# Print the top 10 indices
print(top_10_indices)
```

```python
# Recommending items
for index in top_10_indices:
    if index in prod_dictionary:
        prod_name = prod_dictionary[index]
        print(f"Product {index}: {prod_name}")
    else:
        print(f"Product {index}: Not found in dictionary")
```

```python
# Finding the corresponding product_id from the original DataFrame based on the given dow and hour
def find_matching_product(prod_dow_hour, dow, hour):
    matching_products = []

    for index, row in prod_dow_hour.iterrows():
        order_dow = row['order_dow']
        order_hour_of_day = row['order_hour_of_day']
        product_id = row['product_id']

        if order_dow == dow and order_hour_of_day == hour:
            matching_products.append(product_id)

    return matching_products
```

```python
def evaluate_recommendations(true_items, recommended_items, K):
    # Precision@K
    precision = len(set(true_items) & set(recommended_items[:K])) / float(K)
    
    # Recall@K
    recall = len(set(true_items) & set(recommended_items[:K])) / float(len(true_items))
    
    # nDCG@K
    dcg = 0.0
    idcg = 0.0
    for i in range(K):
        item = recommended_items[i]
        if item in true_items:
            relevance = 1.0
        else:
            relevance = 0.0
        dcg += (2 ** relevance - 1) / np.log2(i + 2)  # i+2 to start indexing from 1
        idcg += (2 ** 1 - 1) / np.log2(i + 2)
    ndcg = dcg / idcg
    
    # Item coverage@K
    item_coverage = len(set(recommended_items[:K])) / float(len(set(recommended_items)))
    
    # MAP@K
    avg_precision = 0.0
    num_hits = 0
    for i in range(K):
        item = recommended_items[i]
        if item in true_items:
            num_hits += 1
            avg_precision += num_hits / float(i + 1)
    if num_hits > 0:
        avg_precision /= num_hits
    mean_avg_precision = avg_precision / min(len(true_items), K)
    
    return precision, recall, ndcg, item_coverage, mean_avg_precision
```

```python
print("precision@K, recall@K, ndcg@K, item_coverage@K, MAP@K are: ")
evaluate_recommendations(original, top_10_indices, 7)
```
